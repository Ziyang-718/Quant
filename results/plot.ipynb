### Plots

```python
import matplotlib.pyplot as plt

# Data
loss = [
    8.976303100585938,
    8.714662551879883,
    8.334434509277344,
    8.024924278259277,
    7.7142014503479
]
acc = [
    0.10037339478731155,
    0.09990517795085907,
    0.11446318030357361,
    0.13534969091415405,
    0.13662222027778625
]
epochs = list(range(1, len(loss) + 1))

# Plot
fig, ax1 = plt.subplots()

# Loss plot
color = 'tab:red'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color=color)
ax1.plot(epochs, loss, color=color, marker='o', label='Loss')
ax1.tick_params(axis='y', labelcolor=color)

# Accuracy plot (twin axis)
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Accuracy', color=color)
ax2.plot(epochs, acc, color=color, marker='x', label='Accuracy')
ax2.tick_params(axis='y', labelcolor=color)

# Title and layout
plt.title('Training Loss and Accuracy per Epoch')
fig.tight_layout()
plt.show()
```

### Scratchpad

```python
import json, glob

maxlen = 512

def corpus():
    """语料生成器
    """
    while True:
        p = 'data/test.json'
        for f in sorted(glob.glob(p)):
            with open(f, errors='ignore') as f:
                for l in f:
                    l = json.loads(l)
                    yield l['content'][:int(maxlen * 1.2)]

generator = corpus()
```
